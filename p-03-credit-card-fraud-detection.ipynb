```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report
```

# Reading Data

```python
data = pd.read_csv("/kaggle/input/creditcardfraud/creditcard.csv")
data.head()
```

# Removing Duplicates

```python
data.drop_duplicates(inplace = True)
```

```python
data.duplicated().sum()
```

```python
data.describe()
```

```python
cor = data.corr()

cor_target = abs(cor["Class"])
#Selecting highly correlated features
relevant_features = cor_target[cor_target>0.05]
relevant_features
```

```python
data["Class"].value_counts()
```

```python
X = data.drop('Class', axis = 1)
Y = data['Class']
```

# Balancing Classes with SMOTE technique

```python
smote = SMOTE(random_state = 42)
X_balanced, y_balanced = smote.fit_resample(X, Y)
```

```python
y_balanced.value_counts()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_balanced, y_balanced, test_size = 0.3, random_state = 42) 
```

# Scaling Data

```python
sc = StandardScaler()
X_train_scaled = sc.fit_transform(X_train)
X_test_scaled = sc.transform(X_test)
```

# Applying Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression(solver='liblinear', random_state=0)

logreg.fit(X_train_scaled, y_train)
```

```python
y_pred = logreg.predict(X_test_scaled)
```

# Average Test Accuracy

```python
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator =logreg, X = X_test_scaled, y = y_test, cv = 10)
print("Accuracy: {:.2f} %".format(accuracies.mean()*100))
```

# Average Train Accuracy

```python
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator =logreg, X = X_train_scaled, y = y_train, cv = 10)
print("Accuracy: {:.2f} %".format(accuracies.mean()*100))
```

# Confusion Matrix

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)

sns.heatmap(cm,annot=True, fmt='d', cmap='YlGnBu')
plt.xlabel("Predicted")
plt.ylabel("Actual ")
plt.title("Confusion Matrix")
```
